```python
from datetime import datetime


def create_account():
    name = input("Enter your name: ")
    account_number = input("Enter your account number: ")
    account = {
        "name": name,
        "account_number": account_number,
        "balance": 0,
        "transaction_history": [],
        "transaction_Count":1
    }
    return account

def deposit(account, amount):
    if amount > 0:
        account["balance"] += amount
        account["transaction_history"].append(f"{account['transaction_Count']}. Credited PKR {amount} on {datetime.now().strftime('%d/%m/%Y %H:%M:%S')}")
        print(f"Dear {account['name']}, PKR {amount} has been credited to your account. Your Current balance is: {account['balance']}.")
        account['transaction_Count'] += 1
    else:
        print("Deposit amount must be positive.")

def withdraw(account, amount):
    if amount > 0:
        if amount <= account["balance"]:
            account["balance"] -= amount
            account["transaction_history"].append(f"{account['transaction_Count']}. Debited  PKR {amount} on {datetime.now().strftime('%d/%m/%Y %H:%M:%S')}")
            print(f"Dear {account['name']}, PKR {amount} has been debited from your account. Your Current balance is: {account['balance']}.")
            account['transaction_Count'] += 1
        else:
            print(f"Dear {account['name']}, you have insufficient funds for this transaction. Please deposit money in order to proceed.")
    else:
        print("Please enter amount greater than 0")

def check_balance(account):
    print(f"Current balance: {account['balance']}")

def view_transactions(account):

    if account["transaction_history"]:
        print(f"Dear {account['name']}, your Transaction History is:")
        for transaction in account["transaction_history"]:
            print(transaction)
    else:
        print("No transactions found")


#-------------------------------------------------------------------------------------
account = None

while True:
    print("*******************************************************************")
    print("************************Welcome to DigiTank************************")
    print("*******************************************************************")
    print("Hello Customer. Here are the desired services which you can avail: ")
    print("1 - Account Opening")
    print("2 - Deposit Money")
    print("3 - Withdraw money")
    print("4 - Check Account Balance")
    print("5 - View Transaction History")
    print("6 - Exit")
    choice = int(input("Please type an option number for your desired service: "))
    print()
    match choice:
      case 1:
        account = create_account()
        print(f"Account created successfully! Welcome {account['name']}.")
        print()

      case 2:
        if account:
          amount = float(input("Enter amount to deposit: "))
          deposit(account, amount)
          print()
        else:
          print("Please create an account first.")
          print()

      case 3:
        if account:
          amount = float(input("Enter amount to withdraw: "))
          withdraw(account, amount)
          print()
        else:
          print("Please create an account first.")
          print()

      case 4:
        if account:
          check_balance(account)
          print()
        else:
          print("Please create an account first.")
          print()
      case 5:
        if account:
          view_transactions(account)
          print()
        else:
          print("Please create an account first.")
          print()
      case 6:
        print("Thank you for choosing Bank DigiTank!")
        break

      case _:
        print("Invalid option chosen. Please choose a valid option to proceed.")


```
